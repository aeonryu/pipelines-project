```python
url='https://science.sciencemag.org/content/359/6371'
result=requests.get(url).content
soup=BeautifulSoup(result,'html5')

tabla = []
for a in soup.find_all('div', class_='media__body'):
    tabla.append(a.text.strip().split("\n")) 
tabla
```

```python
columnas = ['Title', 'Authors','Date','Extra','Mini Abstract','Kind of text','Full/Abstract','a','b','c']
df = pd.DataFrame(tabla, columns = columnas)
df
```
